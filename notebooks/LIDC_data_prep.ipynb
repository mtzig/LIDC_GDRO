```python
import pandas as pd
import numpy as np
import os
```

```python
df_LIDC = pd.read_csv('../data/LIDC_labels.csv')
df_LIDC.head()
```

```python
df_labels = df_LIDC[['noduleID', 'spiculation', 'malignancy']]
df_labels.head()
```

```python
df_labels['Spiculation_b'] = np.where()
```

# Get Nodule ID

```python
image_folder = '../LIDC(MaxSlices)_Nodules'
noduleIDs = []

for file in os.scandir(image_folder):
    noduleID = int(file.name.split('.')[0])
    noduleIDs.append(noduleID)
```

```python
len(noduleIDs)
```

```python
df_labels = df_labels[df_labels['noduleID'].isin(noduleIDs)]
```

```python
df_labels.to_csv('../data/LIDC_labels.csv')
```
