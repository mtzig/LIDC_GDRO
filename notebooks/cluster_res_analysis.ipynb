```python
all_arrays = []
for i in range(len(cluster_all)):
    all_arrays.append(cluster_all.iloc[i].values)
```

```python
#https://stackoverflow.com/questions/38504737/calculate-percentage-of-count-for-a-list-of-arrays

import numpy_indexed as npi
arrs = all_arrays
idx = [np.ones(len(a))*i for i, a in enumerate(arrs)]
(rows, cols), table = npi.count_table(np.concatenate(idx), np.concatenate(arrs))
table = table / table.sum(axis=1, keepdims=True)
print(table * 100)
```

```python
max_percentage = []

for i in range(len(table)):
    max_percentage.append(max(table[i]))
```
